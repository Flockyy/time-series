```python
kept_features = ['fields.gc_obo_type_c', 'fields.gc_obo_nature_c', 'fields.date']
```

```python
dfeurope2022 = pd.json_normalize(europe2022.json()['records'])
dfeurope2022 = dfeurope2022[kept_features]

dfeurope2021 = pd.json_normalize(europe2021.json()['records'])
dfeurope2021 = dfeurope2021[kept_features]

dfeurope2020 = pd.json_normalize(europe2020.json()['records'])
dfeurope2020 = dfeurope2020[kept_features]

dfeurope2019 = pd.json_normalize(europe2019.json()['records'])
dfeurope2019 = dfeurope2019[kept_features]

dfeurope2018 = pd.json_normalize(europe2018.json()['records'])
dfeurope2018 = dfeurope2018[kept_features]

dfeurope2017 = pd.json_normalize(europe2017.json()['records'])
dfeurope2017 = dfeurope2017[kept_features]

dfeurope2016 = pd.json_normalize(europe2016.json()['records'])
dfeurope2016 = dfeurope2016[kept_features]

# dfeurope2015 = pd.json_normalize(europe2015.json()['records'])
# dfeurope2015 = dfeurope2015[kept_features]

# dfeurope2014 = pd.json_normalize(europe2014.json()['records'])
# dfeurope2014 = dfeurope2014[kept_features]

# dfeurope2013 = pd.json_normalize(europe2013.json()['records'])
# dfeurope2013 = dfeurope2013[kept_features]
```

```python
# alltimedf = dfeurope2013.merge(dfeurope2014, how='outer')
# alltimedf = alltimedf.merge(dfeurope2015, how='outer')
# alltimedf = alltimedf.merge(dfeurope2016, how='outer')
alltimedf = dfeurope2016.merge(dfeurope2017, how='outer')
alltimedf = alltimedf.merge(dfeurope2018, how='outer')
alltimedf = alltimedf.merge(dfeurope2019, how='outer')
alltimedf = alltimedf.merge(dfeurope2020, how='outer')
alltimedf = alltimedf.merge(dfeurope2021, how='outer')
alltimedf = alltimedf.merge(dfeurope2022, how='outer')
```

```python
alltimedf['date'] = pd.to_datetime(alltimedf['fields.date'])
```

```python
alltimedf['month'] = alltimedf.date.dt.month
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

a = alltimedf.groupby("month").size().values
df = alltimedf.drop_duplicates(subset="month").assign(Count=a)
df['mean'] = df.mean(axis=1)
df
```

```python
alltimedf['count'] = 1
res = alltimedf.groupby(alltimedf['date'].dt.strftime('%Y %d'))['count'].sum().reset_index()
sns.lineplot(data = res, x=res.date, y = res['count'])
```
