```python
import pandas as pd
df  = alltimedf.copy()
df.reset_index(inplace=True)
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
```

```python
df_agg = df.groupby('date')['count'].sum().reset_index()
df_agg = df_agg.set_index('date')
df_agg
```

```python
y = df_agg['count'].resample('D').sum()
```

```python
y
```

```python
import statsmodels.api as sm
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
decomposition.plot()
```

* Est ce que la time-série est stationnaire? 

Non, on peut observer une tendance à la baisse dans le graphique si dessus

* Est ce que la time-série présente une tendance?

Oui, une tendance à la baisse puis à la hausse

* Est ce que la time-série présente une saisonnalité?

Oui, une saisonalité annuel

* Est ce que la time série voit sa variance varier?

Oui, sa variance diminue

```python
df_pp = df.copy().reset_index()
df_pp = df_pp[['date', 'count']]
df_pp['date'] = df_pp['date'].dt.date
df_pp = df_pp.rename(columns={'date': 'ds', 'count': 'y'})
df_pp = df_pp.groupby(['ds'])['y'].sum().reset_index()
df_pp
```

```python
from prophet import Prophet
m = Prophet()
m.fit(df_pp)
future = m.make_future_dataframe(50, freq='MS')
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
from prophet import Prophet
m = Prophet(seasonality_mode='multiplicative')
m.fit(df_pp)
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
fig = m.plot_components(forecast)
```

```python
m = Prophet(seasonality_mode='multiplicative')
m.add_seasonality('quarterly', period=91.25, fourier_order=8, mode='additive')
m.add_regressor('regressor', mode='additive')
```

```python
m = Prophet(mcmc_samples=300)
forecast = m.fit(df_pp, show_progress=False).predict(future)
```
